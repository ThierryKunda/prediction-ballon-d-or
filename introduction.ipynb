### Colonnes à pourcentages

```python
# On récupère la liste des colonnes dont le nom ne contient pas '%'
cols = [name for name in list(dataset.columns.values) if '%' not in name]
print(cols)
```

```python
dataset = dataset.filter(items=cols)
dataset.shape
```
