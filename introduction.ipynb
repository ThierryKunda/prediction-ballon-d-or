# Encodage des données

```python
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
from pandas import DataFrame
def get_cat_values(data: DataFrame, feature_name: str) -> list[str]:
    """Récupère les valeurs catégorielles à partir du nom de la colonne après un encodage 1-hot.\n
    En cas de encodage à k-1 valeurs, la valeur non explicite ne sera pas récupérée
    """
    values: list[str] = list(data.filter(regex=feature_name+"*").columns.values)
    return [v.replace(feature_name+"_", "") for v in values]
```

```python
# Position du joueur
get_cat_values(dataset, "Pos")
```

```python
# Championnat du club du joueur
get_cat_values(dataset, "Comp")
```

# Visualisation

## Valeurs catégorielles

```python
# On rappelle la liste des features
print(dataset.columns.values, f"Nombre de features : {len(dataset.columns.values)}", sep="\n")
```

Il y en a beaucoup...

```python
ligues = (len(dataset[dataset['Comp_La Liga'] == 1]),
len(dataset[dataset['Comp_Ligue 1'] == 1]),
len(dataset[dataset['Comp_Premier League'] == 1]),
len(dataset[dataset['Comp_Serie A'] == 1]))
ligues
```

### Répartition des joueurs dans les 5 grands championnats

```python
total_joueurs = dataset.shape[0]
nb_par_champ = dict(zip(get_cat_values(dataset, "Comp"), ligues))
# On ajoute la Bundesliga, retiré après le 1-hot
nb_par_champ["Bundesliga"] = total_joueurs - sum(ligues)
nb_par_champ
```

```python
fig, ax = plt.subplots()
# Taille de l'image
fig.set_size_inches(7,7)
# Etiquettes avec associations nom de la ligue - pourcentage du total (après nettoyage et encodage)
perc_labels = [f"{cmp} : {round(nb_par_champ[cmp]/total_joueurs*100, 2)}%" for cmp in nb_par_champ]
# Diagramme circulaire
ax.pie(nb_par_champ.values(), frame=True, center=(4,4), radius=2, labels=perc_labels)
# Paramétrages des axes
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
plt.show()
```

On peut comparer avec le dataset d'origine

```python
filename = 'dataset.csv'
df_orig = pd.read_csv(filename, sep=";", header=0)
champ_names = ["Premier League", "Ligue 1", "La Liga", "Bundesliga", "Serie A"]
nbpc = {cmp: len(df_orig[df_orig['Comp'] == cmp]) for cmp in champ_names}

fig, ax = plt.subplots()
# Taille de l'image
fig.set_size_inches(7,7)
# Diagramme circulaire
ax.pie(nb_par_champ.values(), frame=True, center=(4,4), radius=2, labels=[f"{cmp} : {round(nbpc[cmp]/total_joueurs*100, 2)}%" for cmp in nbpc])
# Paramétrages des axes
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
plt.show()
```

```python
# Affichage des joueurs retirés par championnats
for k1 in nb_par_champ:
    print(f"{k1} : {nbpc[k1] - nb_par_champ[k1]}")
```

On remarque bien que l'on a retiré les valeurs de manière équilibrée par rapport à la répartition entre chaque championnat

### Répartition des joueurs au niveau des postes

```python
postes_prev = set(df_orig['Pos'])
postes_prev
```

```python
noms_postes = list(dataset.filter(regex='Pos_*', axis=1).columns.values)
postes = {nom.replace("Pos_", ""): len(dataset[dataset[nom] == 1]) for nom in noms_postes}
postes["DF"] = dataset.shape[0] -  sum(postes.values())
postes
```

On remarque déjà qu'il n'y a qu'une seule occurrence pour le poste de GKMF (GoalKeeper Midfielder ?).
On le supprimera plus tard dans le dataset car trop rare.

```python
fig, ax = plt.subplots()
fig.set_size_inches(12,12)
perc_labels = [f"{p} : {round(postes[p]/dataset.shape[0]*100, 2)}%" for p in postes]
ax.pie(postes.values(), frame=True, center=(4,4), radius=2, labels=perc_labels)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
plt.show()
```

On remarque plusieurs choses :
- le poste de défenseur (`DF`) est prédominante, suivi de milieu de terrain (`MF`) et de arrière latéral-défenseur latéral
- il y a des mélanges de postes pour indiquer (on imagine) que le joueur peut occuper/a déjà occupé plusieurs postes ou rôles (voire de manière fréquente)
