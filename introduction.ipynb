# Encodage des données

```python
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
from pandas import DataFrame
def get_cat_values(data: DataFrame, feature_name: str) -> list[str]:
    """Récupère les valeurs catégorielles à partir du nom de la colonne après un encodage 1-hot.\n
    En cas de encodage à k-1 valeurs, la valeur non explicite ne sera pas récupérée
    """
    values: list[str] = list(data.filter(regex=feature_name+"*").columns.values)
    return [v.replace(feature_name+"_", "") for v in values]
```

```python
# Position du joueur
get_cat_values(dataset, "Pos")
```

```python
# Championnat du club du joueur
get_cat_values(dataset, "Comp")
```

### Répartition des joueurs dans les 5 grands championnats

```python
total_joueurs = dataset.shape[0]
nb_par_champ = dict(zip(get_cat_values(dataset, "Comp"), ligues))
# On ajoute la Bundesliga, retiré après le 1-hot
nb_par_champ["Bundesliga"] = total_joueurs - sum(ligues)
nb_par_champ
```

```python
fig, ax = plt.subplots()
ax.pie(nb_par_champ.values(), frame=True, center=(4,4), radius=2)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
plt.show()
```
