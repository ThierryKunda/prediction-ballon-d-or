### Rééquilibrage du jeu de données

Pour rééquilibrer le jeu de données, on va dupliquer de façon homogène les lignes
jusqu'à avoir à peu près le même nombre de joueurs à chaque poste,
en prenant le poste majoritaire comme taille référence.

```python
# Nombre total de joueurs
dataset.shape[0]
```

```python
# "Tailles" relatives des "DF" par rapport aux autres postes
ratios = {p: round(postes["DF"]/postes[p]) for p in postes if p != "DF"}
ratios
```

Soit $p$ un poste (autre que `DF`), à partir de $r = \frac{|p|}{|poste_{DF}|}$, on sait que la nouvelle taille du poste $p$ est $p\leftarrow r*p$.

```python
def dupl_homogene(dataset: DataFrame, colonne_maj: str, ratios: dict[str, int]) -> DataFrame:
    dupl: list[Series] = []
    for p in ratios:
        for _ in range(ratios[p]):
            df = dataset[dataset['Pos'] == p].copy(deep=True)
            # print(df)
            # print(list(ligne))
            dupl.append(df)
    dupl.append(dataset[dataset['Pos'] == colonne_maj])
    # print(dupl)
    return pd.concat(dupl, axis=0)
```

```python
def get_cat_values(data: DataFrame, feature_name: str) -> list[str]:
    """Récupère les valeurs catégorielles à partir du nom de la colonne après un encodage 1-hot.\n
    En cas de encodage à k-1 valeurs, la valeur non explicite ne sera pas récupérée
    """
    feature = []
    for name in data.columns.values:
        if name.startswith(feature_name):
            feature.append(name)
    return [ft.replace(feature_name+"_", "") for ft in feature]
```

# Système de score
Plutôt que de directement partir du jeu de données pour réaliser la tâche que l'on a définit au départ,
on va plutôt définir un système de score qui associera à chaque joueur un score sur lequel l'apprentissage sera appliqué.
Ce score sera défini par un ensemble nouveau de colonnes qui correspond aux features auxquels on a associé un poids.

Par exemple : en considérant les buts comme très positives à l'appréciation du joueur et les fautes comme pénalisants, on peut ainsi
appliquer les calculs suivants :
- score des buts = 3 * le nombre de but
- score des fautes = -2 * le nombre de fautes

## Definition de la fonction de poids

```python
from typing import Callable, Union, Literal
from math import log10, sqrt
```

```python
def nombre_a_la_puissance_10(x: float, seuil_puiss_10: int):
    puiss_diff = 0
    if x != 0:
        puiss_diff = seuil_puiss_10 - round(log10(abs(x)))
    return x * (10**puiss_diff)
```

```python
def genere_poids(
        dataset: DataFrame, poids_map: dict[str, Callable[[Union[int, float]], float]],
        defaut: Callable[[Union[int,float]], float] = lambda _: 0,
        seuil_puiss_10: int = 1,
        supprime_non_def: bool = False,
        trie: bool = False
        ) -> DataFrame:
    """Renvoie le jeu de données avec des colonnes scores pour chaque feature."""
    df = dataset.copy(deep=True)
    for ft in poids_map:
        df[ft] = df[ft].apply(poids_map[ft])
        df[ft] = df[ft].apply(lambda x: nombre_a_la_puissance_10(x, seuil_puiss_10))
    # Poids non définis -> appliquer la fonction de défaut
    autres = set(dataset) - set(poids_map)
    for ft in autres:
        df[ft] = df[ft].apply(defaut)
        df[ft] = df[ft].apply(lambda x: nombre_a_la_puissance_10(x, seuil_puiss_10))
    # Suppression des poids non définis (selon le paramètre supprime_non_def)
    if supprime_non_def:
        df = df.drop(list(autres), axis=1)
    df.insert(0, "Total", df.sum(axis=1))
    if trie:
        return df.sort_values(by=['Total'], ascending=False)
    return df
```

```python
def podium_pourcent(dataset: DataFrame, taille_podium: int) -> DataFrame:
    df = dataset.copy(deep=True)
    classement = []
    # Nouvelle colonne "classement" : [0, 0, ..., 0, 1, 1, ..., 1, ..., taille_podium-1]
    nb_par_podium = dataset.shape[0] // taille_podium
    for i in range(nb_par_podium):
        classement += [i] * taille_podium
    reste_indice_debut = classement[-1] * nb_par_podium
    reste = dataset.shape[0] % taille_podium
    classement += ([reste_indice_debut] * reste)
    df.insert(0, "Classement", classement)
    # On remet les lignes dans l'ordre de l'index -> ordre du début
    df.sort_index(axis=0, inplace=True)
    return df
```

# Entrainement sans pré-traitement

```python
from sklearn.cluster import KMeans
from sklearn.metrics import completeness_score, homogeneity_score
from sklearn.model_selection import train_test_split
```

```python
def train_valid_test_split(
        dataset: DataFrame, test_label: str, test_size: float,
        validation_size: float, rand_states: dict[Literal['validation', 'test'], float] = None
        ) -> tuple[DataFrame, DataFrame, DataFrame, Series, Series, Series]:
    if not rand_states:
        rand_states = {"validation": None, "test": None}
    # Premier séparation autre - test
    X_train, X_test, y_train, y_test = train_test_split(
        dataset.drop(test_label, axis=1),
        dataset[test_label],
        test_size=test_size,
        random_state=rand_states["test"]
    )
    # 2è séparation entrainement - validation
    X_train, X_val, y_train, y_val = train_test_split(
        X_train,
        y_train,
        # Remarque : valid_sur_autre = valid_tot / autre
        test_size=validation_size/(1-test_size),
        random_state=rand_states["validation"]
    )
    return X_train, X_val, X_test, y_train, y_val, y_test
```

```python
# rnd = {"validation": 25, 'test': 50}
X_train, X_val, _, y_train, y_val, _ = train_valid_test_split(dataset_classement, "Classement", 0.1, 0.4)
# y_train, y_val, y_test
```

```python
# On veut des clusters avec une taille pré-définie
kmeans = KMeans(len(X_train) // TAILLE_PODIUM, n_init='auto', random_state=40).fit(X_train)
# kmeans.labels_
```

```python
y_pred = kmeans.predict(X_val)
```

```python
# Vérifie que les points que les points de données de même classe appartiennent au même cluster
completeness_score(y_val, y_pred)
```

```python
homogeneity_score(y_val, y_pred)
```

On arrive à des scores autour de 0,6 et 0,8.
