## Encodage 1-hot avec suppression de la première catégorielle

```python
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
def get_cat_values(data: DataFrame, feature_name: str) -> list[str]:
    """Récupère les valeurs catégorielles à partir du nom de la colonne après un encodage 1-hot.\n
    En cas de encodage à k-1 valeurs, la valeur non explicite ne sera pas récupérée
    """
    feature = []
    for name in data.columns.values:
        if name.startswith(feature_name):
            feature.append(name)
    return [ft.replace(feature_name+"_", "") for ft in feature]
```

# Système de score
Plutôt que de directement partir du jeu de données pour réaliser la tâche que l'on a définit au départ,
on va plutôt définir un système de score qui associera à chaque joueur un score sur lequel l'apprentissage sera appliqué.
Ce score sera défini par un ensemble nouveau de colonnes qui correspond aux features auxquels on a associé un poids.

Par exemple : en considérant les buts comme très positives à l'appréciation du joueur et les fautes comme pénalisants, on peut ainsi
appliquer les calculs suivants :
- score des buts = 3 * le nombre de but
- score des fautes = -2 * le nombre de fautes

## Definition de la fonction de poids

```python
from typing import Callable, Union
from math import log10, sqrt
```

```python
def nombre_a_la_puissance_10(x: float, seuil_puiss_10: int):
    puiss_diff = 0
    if x != 0:
        puiss_diff = seuil_puiss_10 - round(log10(abs(x)))
    return x * (10**puiss_diff)
```

```python
def genere_poids(
        dataset: DataFrame, poids_map: dict[str, Callable[[Union[int, float]], float]],
        defaut: Callable[[Union[int,float]], float] = lambda _: 0,
        seuil_puiss_10: int = 1,
        supprime_non_def: bool = False,
        trie: bool = False
        ) -> DataFrame:
    """Renvoie le jeu de données avec des colonnes scores pour chaque feature."""
    df = dataset.copy(deep=True)
    for ft in poids_map:
        df[ft] = df[ft].apply(poids_map[ft])
        df[ft] = df[ft].apply(lambda x: nombre_a_la_puissance_10(x, seuil_puiss_10))
    # Poids non définis -> appliquer de la fonction de défaut
    autres = set(dataset) - set(poids_map)
    for ft in autres:
        df[ft] = df[ft].apply(defaut)
        df[ft] = df[ft].apply(lambda x: nombre_a_la_puissance_10(x, seuil_puiss_10))
    if supprime_non_def:
        df = df.drop(list(autres), axis=1)
        # print(df.sum(axis=1))
        df.insert(0, "Total", df.sum(axis=1))
    if trie:
        return df.sort_values(by=['Total'], ascending=False)
    return df
```

```python
PM = {
    "Goals": lambda x: 3 * x,
    "Assists": lambda x: 2 * x,
    "Blocks": lambda x: x,
    # "Fls": lambda x: -x/2,
    # "PasCmp": lambda x: x/5,
    # "GCA": lambda x: x/4
}
```

```python
# Initialisation du dataset de points
dataset_score = genere_poids(dataset, PM, supprime_non_def=True, trie=True)
dataset_score
```

```python
def topX(dataset: DataFrame, taille_podium: int) -> DataFrame:
    df = dataset.copy(deep=True)
    # Nouvelle colonne "classement" : [5, 5, ..., 5, 10, 10, ..., 10, ..., 100, 100, ..., 100]
    classement = [[v] * taille_podium for v in range(taille_podium, df.shape[0], taille_podium)]
    reste_indice_debut = classement[-1][0] + 1
    taille_reste = (dataset.shape[0]-reste_indice_debut+1)
    reste = [np.int16(reste_indice_debut)] * taille_reste 
    classement.append(reste)
    classement = [np.int16(element) for podium in classement for element in podium]
    df.insert(0, "Classement", classement)
    return df
```

# Entrainement sans pré-traitement

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
```

```python
def train_valid_test_split(
        dataset: DataFrame, test_label: str, test_size: float,
        validation_size: float, rand_states: dict[str, float] = None
        ) -> tuple[DataFrame, DataFrame, DataFrame, Series, Series, Series]:
    # Premier séparation autre - test
    X_train, X_test, y_train, y_test = train_test_split(
        dataset.drop(test_label, axis=1),
        dataset[test_label],
        test_size=test_size,
        random_state=rand_states["test"]
    )
    # 2è séparation entrainement - validation
    X_train, X_val, y_train, y_val = train_test_split(
        X_train,
        y_train,
        # Remarque : valid_sur_autre = valid_tot / autre
        test_size=validation_size/(1-test_size),
        random_state=rand_states["validation"]
    )
    return X_train, X_val, X_test, y_train, y_val, y_test
```

```python
X_train, X_val, X_test, y_train, y_val, y_test = train_valid_test_split(dataset_classement, "Classement", 0.1, 0.4, {"validation": 25, "test": 50})
y_train, y_val, y_test
```
